```python
import pandas as pd

dp = pd.read_csv('C:/Users/Propietario/Desktop/Jupyter/2. Pandas/Datasets/presidents.csv')
dp.head()
```

```python
dp["First"]=dp['President'] #Copy of the first column
# Then we can call replace() and just have a pattern that matches the last name and set it to an empty string
dp["First"]=dp["First"].replace("[ ].*", "", regex=True)
# Now let's take a look
dp.head()
```

```python
del dp['First']
dp.head()
```

```python
def splitname(row):
    row['First'] = row['President'].split(' ')[0]
    row['Last'] = row['President'].split(' ')[-1]
    return row

dp = dp.apply(splitname, axis='columns')
dp.head()
```

```python
del dp['First']
del dp['Last']
dp.head()
```

```python
pattern="(^[\w]*)(?:.* )([\w]*$)"


dp["President"].str.extract(pattern).head()
```

```python
pattern="(?P<First>^[\w]*)(?:.* )(?P<Last>[\w]*$)"

# Now call extract
names=dp["President"].str.extract(pattern).head()
names
```

```python
dp["First"]=names["First"]
dp["Last"]=names["Last"]
dp.head()
```

```python
dp["Born"]=dp["Born"].str.extract("([\w]{3} [\w]{1,2}, [\w]{4})")
dp["Born"].head()
```

```python
dp["Born"]=pd.to_datetime(dp["Born"])
dp["Born"].head()
```
